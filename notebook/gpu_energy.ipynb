```python
zero_ratio = (gpu == 0.0).sum() / len(gpu)
cols_to_drop = zero_ratio[zero_ratio > 0.5].index
gpu_cleaned = gpu.drop(columns=cols_to_drop)
gpu_cleaned.shape
```

Burada verileri %50 üstünde boş olan sütunları sildik. Veri setini rahatlatmaya çalışıyorum. 

```python
usable_cols = [col for col in gpu_cleaned.select_dtypes(include=['float64', 'int64']).columns 
               if gpu_cleaned[col].nunique() > 1]

Q1 = gpu_cleaned[usable_cols].quantile(0.25)
Q3 = gpu_cleaned[usable_cols].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outlier_flags = (gpu_cleaned[usable_cols] < lower_bound) | (gpu_cleaned[usable_cols] > upper_bound)
gpu_cleaned["outlier_score"] = outlier_flags.sum(axis=1)
gpu_outlier_sorted = gpu_cleaned.sort_values("outlier_score", ascending=False)

gpu_outlier_sorted.head(10)
```

Burada IQR çalışması yaptık ve en aykırıları üstten aşağı doğru sıraladık. Bu çıktıda en uç değerlerin genelde yüksek segmentteki kartlarda toplandığını gösteriyor. En yüksek performanslı ekran kartları genelde yine aynı şekilde çok enerji tüketiyorlar. IQR'leri outlier_score sütunu ile belirttik en sonda.

```python
hashrate_cols = [col for col in gpu_cleaned.columns if any(unit in col.lower() for unit in ['h/s', 'mh/s', 'kh/s', 'gh/s', 'sol/s', 'mproof/s']) and 'watt' not in col.lower()]
gpu_cleaned["supported_algo_count"] = (gpu_cleaned[hashrate_cols] > 0).sum(axis=1)
```

Bu yeni sütun her ekran kartının kaç farklı algoritmayı desteklediğini gösteriyor.

```python
power_cols = [col for col in gpu_cleaned.columns if 'watt' in col.lower()]
gpu_cleaned["total_power_consumption"] = gpu_cleaned[power_cols].sum(axis=1)
```

Yine bu sütun her ekran kartının tüm algoritmalar için toplamda ne kadar enerji harcadığını gösteriyor. Yani ekran kartının genel güç ihtiyacını tek sayı olarak veriyor. Enerji analizleri için temel metriklerden biridir. İlk sütunda içinde watt sözcüğü geçen sütunları bulduk ikinci satırda yeni özellik ürettik.

```python
coin_power_cols = [col for col in gpu_cleaned.columns 
                   if 'power' in col.lower() and 'watt' in col.lower() and not any(x in col.lower() for x in ['a2000', 'cmp', 'p10', 'vega', 'vii', 'name'])]
gpu_cleaned["coin_energy_total"] = gpu_cleaned[coin_power_cols].sum(axis=1)
```

Burada her satırda algoritmaların (coinlerin) toplam enerji ihtiyacını gösterir. Coin’lerin enerji verimliliği analizleri için temel göstergedir.

```python
gpu_cleaned["algo_efficiency"] = gpu_cleaned["algo_efficiency"].replace([float("inf"), -float("inf")], pd.NA)
gpu_cleaned["algo_efficiency"] = gpu_cleaned["algo_efficiency"].fillna(0)
```

Bu özellik, ekran kartlarının her Watt enerjiyle kaç algoritmayı destekleyebildiğini gösterir. Az enerjiyle çok iş yapan GPU’ları bulmamıza yardımcı olur.

```python
gpu.columns.tolist()
```

```python
algo_columns = [col for col in gpu_cleaned.columns if col.endswith(("(Mh/s)", "(Gh/s)", "(H/s)", "(Sol/s)", "(Kh/s)", "(Gp/s)", "(Mproof/s)"))]
gpu_cleaned["algo_count"] = gpu_cleaned[algo_columns].gt(0).sum(axis=1)
plt.figure(figsize=(14,6))
sns.barplot(
    x=gpu_cleaned.sort_values("algo_count", ascending=False)["Name"],
    y=gpu_cleaned.sort_values("algo_count", ascending=False)["algo_count"]
)
plt.title("Ekran Kartlarının Desteklediği Algoritma Sayısı", fontweight='bold')
plt.xlabel("Ekran Kartı Modeli", fontweight='bold')
plt.ylabel("Algoritma Sayısı", fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Bu grafikte hangi ekran kartının kaç farklı algoritmayı desteklediğini görüyoruz. En çok algoritmayı destekleyen modeller genelde NVIDIA’nın 30 serisi kartları (örneğin 3080TI, 3080, 3070TI). Bu kartlar farklı blockchain sistemleriyle çalışmaya daha uygun. Az algoritma destekleyen kartlar ise genellikle özel amaçlı ya da düşük uyumluluklu modeller. Bu grafik sadece çok yönlülüğü gösteriyor enerji verimliliği gibi diğer performans kriterlerini göstermez.

```python
top_efficiency = gpu_cleaned.sort_values("algo_efficiency", ascending=False).dropna(subset=["algo_efficiency"])

plt.figure(figsize=(14,6))
sns.barplot(
    x=top_efficiency["Name"],
    y=top_efficiency["algo_efficiency"],
    palette="viridis"
)
plt.xticks(rotation=45, ha='right')
plt.xlabel("Ekran Kartı Modeli", fontweight='bold')
plt.ylabel("Enerji Başına Algoritma Verimliliği", fontweight='bold')
plt.title("En Verimli Ekran Kartları", fontweight='bold')
plt.tight_layout()
plt.show()
```

Bu grafikte ekran kartlarının enerji başına algoritma verimliliği karşılaştırılmış. Yani bir watt enerjiye karşılık hangi ekran kartı daha fazla algoritma çalıştırabiliyor sorusunun cevabını veriyor. Grafiğe baktığımızda en soldaki kart en yüksek verimliliğe sahip. Bu, bazı ekran kartlarının çok enerji tüketmesine rağmen yeterince fazla algoritma çalıştıramadığını gösteriyor. En verimli olan model grafikte en üstte yer alan 6400F modeli. Bu model AMD’ye ait bir ekran kartıdır ve genellikle düşük güç tüketimi ile öne çıkar.
Bu verimlilik kıyaslaması, sadece performansa değil aynı zamanda enerjiye de dikkat edilmesi gereken projelerde oldukça kritiktir. Örneğin bir madenciysen, sadece yüksek performans değil enerji maliyetine karşı aldığın kazanç da önemlidir. Bu grafik tam olarak bunu ölçüyor.

```python
top5_efficient = gpu_cleaned.sort_values("algo_efficiency", ascending=False).head(5)

plt.figure(figsize=(10,6))
sns.barplot(data=top5_efficient, x="Name", y="algo_efficiency", palette="viridis")
plt.title("Madencilik İçin En Verimli İlk 5 Ekran Kartı", fontweight='bold')
plt.xlabel("Ekran Kartı Modeli", fontweight='bold')
plt.ylabel("Enerji Başına Algoritma Verimliliği", fontweight='bold')
plt.tight_layout()
plt.show()
```

Bu grafikte enerji başına algoritma verimliliği en yüksek ilk 5 ekran kartı gösteriliyor. Listenin başında AMD Radeon RX 6400 yer alıyor ve bu kart, enerji verimliliği açısından açık ara en iyisi. Onu yine bir AMD kartı olan 6500XT takip ediyor. Üçüncü sırada yer alan GTX 1660, NVIDIA üretimi ve orta segmentte popüler bir kart. Dördüncü sırada yine NVIDIA GeForce GTX 1060 yer alıyor eski nesil olmasına rağmen hâlâ etkili. Beşinci sırada ise Intel Arc A380 var bu Intel’in yeni nesil ekran kartı ailesinden biri. Yani en verimli 5 kartın ikisi AMD, ikisi NVIDIA ve biri Intel üretimi.

Bu 5 kart arasında fiyat/performans karşılaştırması yaptığımızda genellikle 6500XT, piyasa fiyatı açısından en pahalı olan model olarak öne çıkar.
